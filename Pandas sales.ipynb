```python
new_od = pd.merge(od, cus, how='inner', left_on='customer_id', right_index=True)
new_od.tail()
```

```python
od_2016 = new_od.query("order_date >= '2016-01-01' & order_date <= '2016-12-31'")
od_2016.head()
```

```python
grouped_2016 = od_2016.groupby('city')['sales'].count()
grouped_2016.head()
```

```python
top5 = grouped_2016.sort_values(ascending=False).head(5)
print(top5)
```
